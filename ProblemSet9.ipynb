```python
#W9C1
from libdw import sm

class CM(sm.SM):
    def __init__(self):
        self.start_state = 0

    def get_next_values(self,state, inp):
        assert(state in [0,1])
        assert(inp in [5, 10, 20, 50, 100])
        
        if state == 0 and inp == 100:
            next_state = 0
            output = (0, "coke", 0)
        elif state == 0 and inp == 50:
            next_state = 1
            output = (50, '--', 0)
        elif state == 1 and inp == 100:
            next_state = 0
            output = (0, "coke", 50)
        elif state == 1 and inp == 50:
            next_state = 0
            output = (0, "coke", 0)
        else: 
            next_state = state
            if state == 1:
                bal = 50
            elif state == 0:
                bal = 0
            output = (bal, '--', inp)
        
        return next_state, output
```

```python
#W9C2
from libdw import sm

class SimpleAccount(sm.SM):
    def __init__(self, start_deposit=0):
        self.start_state = start_deposit
    
    def get_next_values(self, state, inp):
        assert(type(inp) == float or type(inp) == int)
        assert(type(state) == float or type(state) == int)
        
        if inp < 0 and state < 100:
            next_state = state + inp - 5
        else: 
            next_state = state + inp
        output = next_state
        
        return next_state, output
```

```python
#W9H1
from libdw import sm

class CommentsSM(sm.SM):
    def __init__(self):
        self.start_state = 0
    
    def get_next_values(self, state, inp):
        assert(state in [0,1])
        
        output = None
        
        if state == 0 and inp == '#':
            next_state = 1
            output = inp
        elif state == 1 and inp != '\n':
            next_state = state
            output = inp
        elif state == 1 and inp == '\n':
            next_state = 0
        else: 
            next_state = state
        
        return next_state, output
```

```python
#W9H2
from libdw import sm

class FirstWordSM(sm.SM):
    def __init__(self):
        self.start_state = 2
    
    def get_next_values(self, state, inp):
        assert(state in [0, 1, 2])
        
        if state == 2:
            if inp == ' ':
                next_state = 0
                output = None
            else:
                next_state = state
                output = inp
        elif state == 0:
            if inp == '\n':
                next_state = 1
                output = None
            else:
                next_state = state
                output = None
        elif state == 1:
            if inp == ' ' or inp == '\n':
                next_state = state
                output = None
            else: 
                next_state = 2
                output = inp
                
        return next_state, output
```
